03.2 Apply Butterworth Filtering

```python
def bw_filter(data):
    order = 3
    nyq = 0.5 * 160
    high = 5 / nyq
    b, a = signal.butter(order, high, btype='lowpass')
    low_passed = signal.filtfilt(b, a, data)
    return low_passed
```

```python
def bw_filter_df(df):
    df['ax (m/s^2)'] = bw_filter(df['ax (m/s^2)'])
    df['ay (m/s^2)'] = bw_filter(df['ay (m/s^2)'])
    df['az (m/s^2)'] = bw_filter(df['az (m/s^2)'])
    df['aT (m/s^2)'] = bw_filter(df['aT (m/s^2)'])
    return df
```

04.2 Feature engineering on sample data

```python
#Lists to catch windows
ax_list = []
ay_list = []
az_list = []
aT_list = []
id_list = []
activity_list = []
sex_list = []
sensor_pos_list = []

```

```python
#Prep Data
def standardize_features(df, id, sex, sensor_pos, activity):
    window = 50
    step = 25

    for i in range(0, df.shape[0] - window, step):
        ax_step = df['ax (m/s^2)'].values[i: i + 100]
        ay_step = df['ay (m/s^2)'].values[i: i + 100]
        az_step = df['az (m/s^2)'].values[i: i + 100]
        aT_step = df['aT (m/s^2)'].values[i: i + 100]

        ax_list.append(ax_step)
        ay_list.append(ay_step)
        az_list.append(az_step)
        aT_list.append(aT_step)
        id_list.append(id)
        sex_list.append(sex)
        sensor_pos_list.append(sensor_pos)
        activity_list.append(activity)

```

```python
#Apply Features
def fe_ml_dataset_prep():
    df = pd.DataFrame()

    #mean
    df['ax_mean'] = pd.Series(ax_list).apply(lambda x: x.mean())
    df['ay_mean'] = pd.Series(ay_list).apply(lambda x: x.mean())
    df['az_mean'] = pd.Series(az_list).apply(lambda x: x.mean())
    df['aT_mean'] = pd.Series(aT_list).apply(lambda x: x.mean())

    #absolute difference - mean
    df['ax_ad_mean'] = pd.Series(ax_list).apply(lambda x: np.mean(np.absolute(x - np.mean(x))))
    df['ay_ad_mean'] = pd.Series(ay_list).apply(lambda x: np.mean(np.absolute(x - np.mean(x))))
    df['az_ad_mean'] = pd.Series(az_list).apply(lambda x: np.mean(np.absolute(x - np.mean(x))))
    df['aT_ad_mean'] = pd.Series(aT_list).apply(lambda x: np.mean(np.absolute(x - np.mean(x))))

    #min
    df['ax_min'] = pd.Series(ax_list).apply(lambda x: x.min())
    df['ay_min'] = pd.Series(ay_list).apply(lambda x: x.min())
    df['az_min'] = pd.Series(az_list).apply(lambda x: x.min())
    df['aT_min'] = pd.Series(aT_list).apply(lambda x: x.min())

    #max
    df['ax_max'] = pd.Series(ax_list).apply(lambda x: x.max())
    df['ay_max'] = pd.Series(ay_list).apply(lambda x: x.max())
    df['az_max'] = pd.Series(az_list).apply(lambda x: x.max())
    df['aT_max'] = pd.Series(aT_list).apply(lambda x: x.max())

    #standard deviation
    df['ax_std'] = pd.Series(ax_list).apply(lambda x: x.std())
    df['ay_std'] = pd.Series(ay_list).apply(lambda x: x.std())
    df['az_std'] = pd.Series(az_list).apply(lambda x: x.std())
    df['aT_std'] = pd.Series(aT_list).apply(lambda x: x.std())

    #median
    df['ax_median'] = pd.Series(ax_list).apply(lambda x: np.median(x))
    df['ay_median'] = pd.Series(ay_list).apply(lambda x: np.median(x))
    df['az_median'] = pd.Series(az_list).apply(lambda x: np.median(x))
    df['aT_median'] = pd.Series(aT_list).apply(lambda x: np.median(x))

    #absolute difference - median
    df['ax_ad_median'] = pd.Series(ax_list).apply(lambda x: np.median(np.absolute(x - np.median(x))))
    df['ay_ad_median'] = pd.Series(ay_list).apply(lambda x: np.median(np.absolute(x - np.median(x))))
    df['az_ad_median'] = pd.Series(az_list).apply(lambda x: np.median(np.absolute(x - np.median(x))))
    df['aT_ad_median'] = pd.Series(aT_list).apply(lambda x: np.median(np.absolute(x - np.median(x))))

    #max - min
    df['ax_mm_diff'] = df['ax_max'] - df['ax_min']
    df['ay_mm_diff'] = df['ay_max'] - df['ay_min']
    df['az_mm_diff'] = df['az_max'] - df['az_min']
    df['aT_mm_diff'] = df['aT_max'] - df['aT_min']

    #interquatile range
    df['ax_iqr'] = pd.Series(ax_list).apply(lambda x: np.percentile(x, 75) - np.percentile(x, 25))
    df['ay_iqr'] = pd.Series(ay_list).apply(lambda x: np.percentile(x, 75) - np.percentile(x, 25))
    df['az_iqr'] = pd.Series(az_list).apply(lambda x: np.percentile(x, 75) - np.percentile(x, 25))
    df['aT_iqr'] = pd.Series(aT_list).apply(lambda x: np.percentile(x, 75) - np.percentile(x, 25))

    #pos count
    df['ax_pos_count'] = pd.Series(ax_list).apply(lambda x: np.sum(x > 0))
    df['ay_pos_count'] = pd.Series(ay_list).apply(lambda x: np.sum(x > 0))
    df['az_pos_count'] = pd.Series(az_list).apply(lambda x: np.sum(x > 0))
    df['aT_pos_count'] = pd.Series(aT_list).apply(lambda x: np.sum(x > 0))

    #neg count
    df['ax_neg_count'] = pd.Series(ax_list).apply(lambda x: np.sum(x < 0))
    df['ay_neg_count'] = pd.Series(ay_list).apply(lambda x: np.sum(x < 0))
    df['az_neg_count'] = pd.Series(az_list).apply(lambda x: np.sum(x < 0))
    df['aT_neg_count'] = pd.Series(aT_list).apply(lambda x: np.sum(x < 0))

    #peaks
    df['ax_peaks'] = pd.Series(ax_list).apply(lambda x: np.sum(x > x.mean()))
    df['ay_peaks'] = pd.Series(ay_list).apply(lambda x: np.sum(x > x.mean()))
    df['az_peaks'] = pd.Series(az_list).apply(lambda x: np.sum(x > x.mean()))
    df['aT_peaks'] = pd.Series(aT_list).apply(lambda x: np.sum(x > x.mean()))

    #kurtosis
    df['ax_kurtosis'] = pd.Series(ax_list).apply(lambda x: stats.kurtosis(x))
    df['ay_kurtosis'] = pd.Series(ay_list).apply(lambda x: stats.kurtosis(x))
    df['az_kurtosis'] = pd.Series(az_list).apply(lambda x: stats.kurtosis(x))
    df['aT_kurtosis'] = pd.Series(aT_list).apply(lambda x: stats.kurtosis(x))

    #skewness
    df['ax_skewness'] = pd.Series(ax_list).apply(lambda x: stats.skew(x))
    df['ay_skewness'] = pd.Series(ay_list).apply(lambda x: stats.skew(x))
    df['az_skewness'] = pd.Series(az_list).apply(lambda x: stats.skew(x))
    df['aT_skewness'] = pd.Series(aT_list).apply(lambda x: stats.skew(x))

    #energy
    df['ax_energy'] = pd.Series(ax_list).apply(lambda x: np.sum(x**2)/100)
    df['ay_energy'] = pd.Series(ay_list).apply(lambda x: np.sum(x**2)/100)
    df['az_energy'] = pd.Series(az_list).apply(lambda x: np.sum(x**2)/100)
    df['aT_energy'] = pd.Series(aT_list).apply(lambda x: np.sum(x**2)/100)

    #labels
    df['id'] = id_list
    df['sex'] = sex_list
    df['sensor_pos'] = sensor_pos_list
    df['activity'] = activity_list

    return df

    
```
